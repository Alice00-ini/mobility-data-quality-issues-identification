```python
import pandas as pd
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pickle
import os
import seaborn as sns
from tqdm.auto import tqdm
```

## 集大成！

```python
# Initialize an empty DataFrame for the final resul
#df_a2 = pd.DataFrame(columns=['time_start_hour', 'time_start_hour_count', 'missing_counts', 'missing_percentage'])

# Define the path to the parent directory
parent_directory = '/scratch/work/lyub2/Problem_output_September/'

df_list = []

# Iterate through all 30 folders with tqdm for progress bar
for day in tqdm(range(1, 31), desc="Processing folders"):
    # Generate folder name in the format 'MMDD_output_csv' (e.g., '0901_output_csv')
    folder_name = f"09{str(day).zfill(2)}_output_csv"
    # Define the full path to the csv file
    file_path = os.path.join(parent_directory, folder_name, 'P1A2_missing_values_hour.csv')
    
    # Check if the file exists to avoid errors
    if not os.path.isfile(file_path):
        print(f"File not found: {file_path}")
        continue
    
    # Read the current CSV file
    df_current = pd.read_csv(file_path, usecols= ['time_start_hour', 'time_start_hour_count', 'missing_counts', 'missing_percentage'])

    
    # Append the current dataframe to the list
    df_list.append(df_current)

# Concatenate all dataframes in the list
df = pd.concat(df_list, ignore_index=True)

df_a2 = df.groupby('time_start_hour').agg({
    'time_start_hour_count': 'sum',
    'missing_counts': 'sum'
}).reset_index()

df_a2['missing_percentage'] = round((df_a2['missing_counts'] / df_a2['time_start_hour_count']) * 100, 2)

df_a2.to_csv('/scratch/work/lyub2/Problem_output_September/P1A2/df_a2.csv', index = False)
```
